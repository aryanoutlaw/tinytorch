# Let's begin working with the project "Electricity consumption prediction"

## The data is taken from [kaggle](https://www.kaggle.com/datasets/thedevastator/240000-household-electricity-consumption-records)
**240,000+ Household Electricity Consumption Records for Machine Learning
A Dataset for The Smart Way to Save Energy**
1. **Date**	= The date of the observation. (Date)
2. **Time**	= The time of the observation. (Time)
3. **Global_active_power**	= The total active power consumed by the household (kilowatts). (Numeric)
4. **Global_reactive_power**	= The total reactive power consumed by the household (kilowatts). (Numeric)
5. **Voltage**	The voltage at which the electricity is delivered to the household (volts). (Numeric)
6. **Global_intensity**	The average current intensity delivered to the household (amps). (Numeric)
7. **Sub_metering_1**	The active power consumed by the kitchen (kilowatts). (Numeric)
8. **Sub_metering_2**	The active power consumed by the laundry room (kilowatts). (Numeric)
9. **Sub_metering_3**	The active power consumed by the electric water heater and air conditioner (kilowatts). (Numeric)

### Why we not selecting "Global_reactive_power"
* This power doesn't perform useful work but is necessary for the operation of certain devices.
* Reactive power is not billed directly to consumers but affects the efficiency of the power distribution system.

While, **"Global_active_power"** on other hand represents the actual power consumed by electrical devices in the household to perform useful work, such as lighting, heating, or running appliances. Active power is the "real" power that contributes to energy consumption and is billed by energy providers.

```python
# Convert the 'Date' and 'Time' columns to a single 'date_time' column by combining two columns
data['date_time'] = pd.to_datetime(data['Date']+' '+data['Time'])

# Converting all necessary columns to numeric format
data['Global_active_power'] = pd.to_numeric(data['Global_active_power'], errors='coerce')
data['Voltage'] = pd.to_numeric(data['Voltage'], errors='coerce')
data['Global_intensity'] = pd.to_numeric(data['Global_intensity'], errors='coerce')
data['Sub_metering_1'] = pd.to_numeric(data['Sub_metering_1'], errors='coerce')
data['Sub_metering_2'] = pd.to_numeric(data['Sub_metering_1'], errors='coerce')
data = data.dropna(subset=['Global_active_power'])

data['date_time'] = pd.to_datetime(data['date_time']) 

# create new columns for year, quarter, month and day
data['year'] = data['date_time'].apply(lambda x : x.year)
data['quarter'] = data['date_time'].apply(lambda x : x.quarter)
data['month'] = data['date_time'].apply(lambda x : x.month)
data['day'] = data['date_time'].apply(lambda x : x.day)

# Drop unnecessary quarters
data = data[data["quarter"] != 3]
data = data[data["quarter"] != 4]

# Create new feature that indicates which part of the day
data["Daytime"] = data['date_time'].dt.hour # dt.hour = It extracts the hour from the datetime values.
data['Daytime'] = data['Daytime'].apply(lambda x : x/8) #The goal is to divide the day into three parts
```

```python
data = data.drop(columns=['Date','Time', 'date_time'], axis=1)
data
```

```python
X = data.drop("Global_active_power", axis=1)
y = data['Global_active_power']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print("X_train :",X_train.shape, " ,", "y_train :", y_train.shape)
print("X_test :", X_test.shape, " ,", "y_test :", y_test.shape)
```

### Using RandomForestRegressor for prediction

```python
regr = RandomForestRegressor()
regr.fit(X_train, y_train)
regr_score = regr.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = regr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Score of RandomForestRegressor: ", regr_score)
print("Mean squared error: ",mse)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model = GradientBoostingRegressor()
model.fit(X_train, y_train)
gbRegr_score = model.score(X_test, y_test)
gby_pred = model.predict(X_test)
print(gbRegr_score)
```

```python
mse_gbRegressor = mean_squared_error(y_test, gby_pred)
mse_gbRegressor
```

```python
print("Score of GradientBoostingRegressor: ",gbRegr_score)
print("Mean squared error: ", mse_gbRegressor)
```
